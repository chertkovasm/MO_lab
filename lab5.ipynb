```python
X = data.drop(['diagnosis'], axis=1)
y = data['diagnosis'].map({'B': 0, 'M': 1})
yvals = data['diagnosis']
print(f'X shape: {X.shape} | y shape: {y.shape} ')
```

Стандартизация данных

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
labels = ['B', 'M']
colors = ['green', 'red']
```

# Метод главных компонентов (Principal Component Analysis, PCA)

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)


for i, label in enumerate(['B', 'M']):
    plt.scatter(X_pca[y == i, 0], X_pca[y == i, 1],
                c=colors[i], label=labels[i], alpha=0.5)

plt.title('РСА Визуализация')
plt.legend(loc="best", shadow=False, scatterpoints=1)
plt.xlabel('1-я компонента')
plt.ylabel('2-я компонента')
plt.show()
```

```python
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)


for i, label in enumerate(['B', 'M']):
    plt.scatter(X_pca[y == i, 0], X_pca[y == i, 1],
                c=colors[i], label=labels[i], alpha=0.5)

plt.title('РСА Визуализация (стандартизированные данные)')
plt.legend(loc="best", shadow=False, scatterpoints=1)
plt.xlabel('1-я компонента')
plt.ylabel('2-я компонента')
plt.show()
```

```python
fast_ica = FastICA(n_components=2)
X_ica = fast_ica.fit_transform(X_scaled)


for i, label in enumerate(['B', 'M']):
    plt.scatter(X_ica[y == i, 0], X_ica[y == i, 1],
                c=colors[i], label=labels[i], alpha=0.5)

plt.title('IСА Визуализация (стандартизированные данные)')
plt.legend(loc="best", shadow=False, scatterpoints=1)
plt.xlabel('1-я компонента')
plt.ylabel('2-я компонента')
plt.show()
```

# DBSCAN для кластеризации

```python
dbscan_cluster1 = DBSCAN(eps = 0.5, min_samples = 8)
X_new = dbscan_cluster1.fit(X_scaled)
```

```python
# Number of Clusters
labels=dbscan_cluster1.labels_
N_clus=len(set(labels))-(1 if -1 in labels else 0)
print('Estimated no. of clusters: %d' % N_clus)

# Identify Noise
n_noise = list(dbscan_cluster1.labels_).count(-1)
print('Estimated no. of noise points: %d' % n_noise)

# Calculating v_measure
print('v_measure =', v_measure_score(y, labels))
```

При eps = 0.5, min_samples = 8 не удалось найти ни одного кластера, все точки данных были отмечены как шум, а V-мера указывает на отстутствие хорошей кластеризации.



```python
X_sc = pd.DataFrame(X_scaled)
cols = X_sc.columns[:2]

plt.scatter(X_sc[cols[0]], X_sc[cols[1]],c=labels)
plt.title('DBSCAN')
plt.show()
```

Найдем число eps с помощью правила логтя:

```python
from ipywidgets import interact
from sklearn.neighbors import NearestNeighbors
import sklearn
#возможные метрики расстояния
sklearn.metrics.pairwise.distance_metrics()

@interact(min_samples=(3, 100, 1))
def plot_dbscan(min_samples):
    knn = NearestNeighbors(n_neighbors = min_samples, metric = 'l2')
    knn.fit(X_scaled)

    dists, inds = knn.kneighbors(X_scaled)
    dists_ar = dists[:, 1:min_samples].mean(axis=1)
    dists_ar.sort()

    plt.scatter(x = range(len(dists_ar)), y=dists_ar)
    plt.show()
```

Для min_samples = 4 eps = 4

```python
eps=4
min_samples=4
dbscan_cluster1 = DBSCAN(eps = eps, min_samples = min_samples)
X_new = dbscan_cluster1.fit(X_scaled)

# Number of Clusters
labels=dbscan_cluster1.labels_
N_clus=len(set(labels))-(1 if -1 in labels else 0)
print('Estimated no. of clusters: %d' % N_clus)

# Identify Noise
n_noise = list(dbscan_cluster1.labels_).count(-1)
print('Estimated no. of noise points: %d' % n_noise)

# Calculating v_measure
print('v_measure =', v_measure_score(y, labels))
```

Удалось определить 2 кластера, при это количество шумов уменьшилось, V-мера теперь не равно нулю, но очень мала.

# DBSCAN для определения выбросов

```python
dbscan = DBSCAN(eps = eps, min_samples = min_samples)
print(dbscan)
```

```python
pred = dbscan.fit_predict(X_scaled)
```

```python
anom_index = [i for i in range(len(pred)) if pred[i] == -1]

values = X_sc.loc[anom_index]
```

```python
plt.scatter(X_sc[cols[0]], X_sc[cols[1]],c=dbscan.labels_)
plt.scatter(values[cols[0]], values[cols[1]], color='r', label='Аномалии')
plt.title('DBSCAN')
plt.legend(loc="best", shadow=False, scatterpoints=1)
plt.show()
```
